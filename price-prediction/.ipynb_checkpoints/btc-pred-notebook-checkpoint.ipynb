```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime
from sklearn.preprocessing import MinMaxScaler
from keras import models
from keras.layers import Dense, LSTM, Dropout
```

```python
price_data = pd.read_csv('bitcoin-price/btc.csv', date_parser=True)
#drop NaN data
price_data.dropna(inplace=True)
price_data.head()
```

```python
price_data.shape
#too large -> will arrange to 1 entry/day
```

```python
price_data['date'] = pd.to_datetime(price_data['Timestamp'],unit='s').dt.date
group = price_data.groupby('date')

weighted_price = group['Weighted_Price'].mean()
volume = group['Volume_(BTC)'].sum()
price_data.drop(['Timestamp'],axis=1,inplace=True)

data = {"weighted_price": weighted_price,
       "volume": volume}
df = pd.concat(data, axis=1)
price_data.head()
```

```python
df
```

```python
#creating train and test datasets
train_price_data = df.loc[:datetime.date(year=2020,month=1,day=1)]
test_price_data = df.loc[datetime.date(year=2020,month=1,day=2):]
train_price_data
```

```python
test_price_data
```

```python
train_price_data.shape[0]
```

```python
#sort train and test data -> 60 day intervals
X_train = []
y_train = []
X_test = []
y_test = []
#scale -> (0,1)
scaler = MinMaxScaler()
train_price_data = scaler.fit_transform(train_price_data)
test_price_data = scaler.fit_transform(test_price_data)

for i in range(60, train_price_data.shape[0]):
    X_train.append(train_price_data[i-60:i])
    y_train.append(train_price_data[i,0])
    
#convert to numpy
X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python
#build rnn
rnn = models.Sequential()

rnn.add(LSTM(units=50, activation='relu',return_sequences=True,input_shape = (X_train.shape[1],2)))
rnn.add(Dropout(0.2))

rnn.add(LSTM(units=60, activation='relu',return_sequences=True))
rnn.add(Dropout(0.3))

rnn.add(LSTM(units=80, activation='relu',return_sequences=True))
rnn.add(Dropout(0.4))

rnn.add(LSTM(units=100, activation='relu'))
rnn.add(Dropout(0.5))

rnn.add(Dense(units=1))
```

```python
rnn.compile(optimizer='adam',loss='mean_squared_error')
rnn.fit(X_train, y_train,epochs=30,batch_size=16)
```

```python
for i in range(60, test_price_data.shape[0]):
    X_test.append(test_price_data[i-60:i])
    y_test.append(test_price_data[i,0])

X_test = np.array(X_test)
y_test = np.array(y_test)
```

```python
#test rnn
y_pred = rnn.predict(X_test)
```

```python
#adjust scale of prices
scale = 1/scaler.scale_[0]

y_pred *= scale
y_test *= scale
```

```python
X_test.shape
```

```python
plt.figure(figsize=(16,5))
plt.plot(y_test, color = 'red', label = 'Actual BTC Price')
plt.plot(y_pred, color = 'blue', label = 'Predicted BTC Price')
plt.title('BTC Price Prediction')
plt.ylabel('Price - USD')
plt.xlabel('Time - Days')
plt.show()
```
